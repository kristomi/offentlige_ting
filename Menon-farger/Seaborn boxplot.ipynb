# Eksempel på horisontalt boxplot 

Laster moduler

```python
import pandas as pd
import seaborn as sns
import json
import requests
import matplotlib.pyplot as plt
from collections import OrderedDict

%matplotlib inline
```

Laster inn Menon-fargene fra en fil jeg har laget tidligere, basert på RGB-verdiene Peter la ut i anledning Noun Project

```python
path = "https://raw.githubusercontent.com/kristomi/offentlige_ting/master/Menon-farger/menon_farger.json"
farger = json.loads(requests.get(path).text, object_pairs_hook=OrderedDict)
farger
```

Lager en fargepalett Seaborn kan lese

```python
menon_palette = sns.color_palette(farger.values())
```

Laster inn et vilkårlig demo-datasett

```python
iris = sns.load_dataset('iris')
iris
```

Lager et boxplott som skal være horisontalt, og bruke Menon-fargene

```python
ax = sns.boxplot(data=iris, orient="h", palette=menon_palette)
```
